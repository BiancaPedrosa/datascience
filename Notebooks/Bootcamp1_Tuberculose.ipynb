Análise dos casos de turbeculose no estado de SP entre 2009 - 2018

Importação de bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

Leitura dos dados

```python
#tabela completa

url="https://raw.githubusercontent.com/BiancaPedrosa/datascience/master/dados/tabelautf8a.csv"

dados = pd.read_csv(url, encoding="ISO-8859-1", engine='python', sep=";",thousands=".", decimal=",", skiprows=1,index_col=0)
dados
```

Tratando a ausência de dados

```python
dados = dados.replace('-', np.nan)
dados.head()
```

Conferindo as células sem valor


```python
dados.T.info()
dados.isnull().sum().sum()
```

```python
dados = dados.astype('float64')
dados.head()
```

Selecionando as colunas usáveis

```python
dados_usaveis = dados.loc[:,'2009':'2018']
dados_usaveis
```

```python
anos = ["2009","2010","2011","2012","2013","2014","2015","2016","2017","2018"]
for ano in anos:
  dados["_"+ano] = np.where(dados[ano]==np.nan,  np.nan,(dados[ano]/dados["POP"+ano]))






```

```python
dadospop=dados.loc[:,"_2009":"_2018"]
dadospop
```

```python
#ordenando
dadospop["Total"] = dadospop.sum(axis=1)
ordenado = dadospop.sort_values("Total", ascending=False)
ordenado = ordenado.drop(columns=['Total'])
ordenado.head(10)
```

```python
grafico = ordenado.head(10).T.plot(figsize=(12, 6))
plt.title('Top 10 Casos de tuberculose no estado de SP', fontsize = 20, fontweight = 'bold')

grafico.set_xlabel('Ano', fontsize=12,fontweight = 'bold')
grafico.set_ylabel('Nº de Internação', fontsize=12, fontweight = 'bold')


```
